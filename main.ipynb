```python
import pandas as pd
original_data = pd.read_csv("ideal.csv")
first_part = original_data.iloc[:int(original_data.shape[0]/4)]
second_part = original_data.iloc[int(original_data.shape[0]/4):int(original_data.shape[0]/2)]
third_part = original_data.iloc[int(original_data.shape[0]/2):int(original_data.shape[0]*3/4)]
fourth_part = original_data.iloc[int(original_data.shape[0]*3/4):]
first_part.to_csv("first_part.csv")
second_part.to_csv("second_part.csv")
third_part.to_csv("third_part.csv")
fourth_part.to_csv("fourth_part.csv")
```
